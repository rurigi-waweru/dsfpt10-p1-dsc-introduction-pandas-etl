```python
import pandas as pd

data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'Temperature': [5, 15, 4, 16]
}

df = pd.DataFrame(data)
df
```

```python
pivoted_df = df.pivot(index = 'Date',
                      columns = 'City',
                      values = 'Temperature')
pivoted_df
```
